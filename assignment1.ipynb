Easiest euclidian distance calculation with only 1 line of code so I dont think there are much of a explaination needed in here.

```python
# Function to compute Euclidean distance
def euclidean_distance(point1, point2):
    return np.sqrt(np.sum((point1 - point2) ** 2))
```

(At first I used Counts function which do this job but I do not know if it is allowed so I did not wanted to take any risks.)\
This class is designed to identify the most common class from a list of neighbors, which would be the majority class in the K-Nearest Neighbors (KNN) algorithm.\
It creates a dictionary called class_count to keep track of how many times each class (label) appears among the neighbors.\
It returns the class with the highest count using the max() function. The key=class_count.get ensures that it returns the class with the highest value (count) from the class_count dictionary.

```python
def most_common_class(neighbors):

    class_count = {}
    for neighbor in neighbors:
        class_count[neighbor] = class_count.get(neighbor, 0) + 1

    # If all neighbors do not belong to the same class and classes exist, we take max.
    return max(class_count, key=class_count.get)
```

This is the classic kNN implementation which is used to classify a data point based on the majority label (or class) among its nearest neighbors in a training dataset.  

For each test_point in X_test, the algorithm calculates the distance between the test point and every train_point in X_train.\
The distance is computed using the euclidean_distance function that I showed at the above.\
The I sorted the distances list in ascending order based on the distance so because the nearest neighbours will be at the start of the list.\
After sorting, the algorithm extracts the labels of the k nearest neighbors\
The algorithm uses the "most_common_class" function to determine the majority class.\
The predicted class will be at the "predictions" list

```python

# Implementing KNN algorithm 
def knn(X_train, y_train, X_test, k):
    predictions = []
    
    # iterating over all points to check.
    for test_point in X_test:
        distances = []
        for i, train_point in enumerate(X_train):
            dist = euclidean_distance(test_point, train_point)
            distances.append((dist, y_train.iloc[i]))

        distances.sort(key=lambda x: x[0])

        k_nearest_neighbors = []
        for i in range(min(k, len(distances))):
            dist, label = distances[i]
            k_nearest_neighbors.append(label)

        prediction = most_common_class(k_nearest_neighbors)
        predictions.append(prediction)
    
    return np.array(predictions)

```

This function "calculate_metrics" is used to calculate accuracy, precision, and recall in classification.  

Firstly, I converted "y_true" and "y_pred" to numpy classes because making operations on the elements of the dataset part.\
Then, I calculated accuracy. Accuracy is calculated as the ratio of correct predictions to the total number of instances.\
Created "precision_per_class" and "recall_per_class". Those dictionaries is created to store precision and recall values for each class.\
Calculated True Positive(TP), False Positive(FP), False Negative(FN) values for each classes individually.(True negative(TN) is not needed for precision and recall calculation because precision is calculated as: TP/(TP+FP), Recall is calculated as: TP/(TP+FN))\
Then calculated the precision and recall values and assign it seperately for each class.\
After the calculations, I took the average of the precision and recall values.  

NOTE: I tried the sklearn's accuracy and f1 score metrics and my "calculate_metrics" function is giving exactly the same result and correct result.



```python
def calculate_metrics(y_true, y_pred):
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)

    # (TP + TN)
    total_correct = np.sum(y_true == y_pred)

    
    total_instances = len(y_true)

    # Accuracy calculation
    accuracy = total_correct / total_instances

    # Precision and Recall calculation 
    precision_per_class = {}
    recall_per_class = {}
    classes = np.unique(y_true)
    
    for c in classes:
        TP = np.sum((y_true == c) & (y_pred == c))
        FP = np.sum((y_true != c) & (y_pred == c))
        FN = np.sum((y_true == c) & (y_pred != c))

        # Precision
        precision_per_class[c] = TP / (TP + FP) if (TP + FP) > 0 else 0

        # Recall
        recall_per_class[c] = TP / (TP + FN) if (TP + FN) > 0 else 0

    # Average precision and recall calculation.
    avg_precision = np.mean(list(precision_per_class.values()))
    avg_recall = np.mean(list(recall_per_class.values()))

    return accuracy, avg_precision, avg_recall
```

<div style="font-size: 20px; color: white;">

        Formula for Min-Max Scaling is:   (X - Xmin) / (Xmax - Xmin)
</div>

```python
def min_max_scaling(X):
    X_scaled = X.copy()  
    
    for column in X.columns:
        col_min = X[column].min()  
        col_max = X[column].max()  
        
        X_scaled[column] = (X[column] - col_min) / (col_max - col_min) if col_max != col_min else 0  # All values will be between 0-1.  
                                                                                                     # if col_max = col_min, denominator will be 0
                                                                                                     # and there will be a undefined problem. 
    
    return X_scaled
```

Before implementing weighted kNN algorithm, I will start with scaling all the values between 0-1 using my previously implemented "min_max_scaling" function.

```python
X_train_scaled = min_max_scaling(X_train)
X_test_scaled = min_max_scaling(X_test)
```

After that, I will use 2 different distance calculation functions to derive the weights in weighted kNN.\
(I deleted the exponential distance weight calculation because I don't know how but it gave the similar(not exact but almost the same) result as inverse.)

```python

def inverse_distance_weight(distance):
    return 1 / distance if distance != 0 else 1e5  

#def exponential_distance_weight(distance):
#    return np.exp(-distance) if distance != 0 else 1e5

def polynomial_distance_weight(distance, p=2):
    return 1 / (distance ** p) if distance != 0 else 1e5
```

<div style="font-size: 25px; color: white;">
    <ul>
         Implementing WEIGHTED KNN Algorithm Using Inverse Distance Weight Calculation
</div>

This function is designed to predict class labels using the K-Nearest Neighbors (KNN) algorithm, where neighbors are weighted by their inverse distance to the test point.\

I startedalculating the distances for the each point and then sort the distances so nearest neighbours will be at start.\

The difference between usual kNN and this function is, this time we are not making predictions based on distances. We make predictions based on the inverse of the distances which is 1/distance.\

This method will give the importance of distance more because when distance is so low, weight will be higher but if the distance will be too high, weight will be so small so it wouldn't be effecting our prediction.\

When predicting, we are summing the weights for each class and choosing the class with the highest total weight.

```python
# Changing our previously written knn algorithm with adding weights.
def weighted_knn_inverse_distance(X_train, y_train, X_test, k):
    predictions = []
    
    for test_point in X_test:
        distances = []
        for i, train_point in enumerate(X_train):
            dist = euclidean_distance(test_point, train_point)
            distances.append((dist, y_train.iloc[i]))
        
        # sorting distances as usual
        distances.sort(key=lambda x: x[0])

        k_nearest_neighbors = distances[:k]

        # Now this part is the part that comes weights. 
        class_weights = {}
        for dist, label in k_nearest_neighbors:
            weight = inverse_distance_weight(dist)
            class_weights[label] = class_weights.get(label, 0) + weight
        
        prediction = max(class_weights, key=class_weights.get)
        predictions.append(prediction)
    
    return np.array(predictions)
```

Now, I want to check if applying PCA(Principal Component Analysis) will reduce the error or not. The logic behind using PCA is, kNN is a classification algorithm that have issues with dimentions. The lower the dimention is, the better result it must give. But keeping too few components may result in loss of important information, hence keeping too high components may defeat the purpose of dimensionality reduction. So as chosing the k value, chosing the number of components are important. 

Starting to apply PCA:

Normalizing data using min-max scaling ensures that all feature values are scaled between 0-1.

```python
# Normalizing data using min-max scaling that I previously defined
X_train_normalized = min_max_scaling(X_train)
X_test_normalized = min_max_scaling(X_test)
```

At first, the mean of the training dataset is calculated, and this mean is subtracted from each feature to center the data.

The covariance matrix of the centered data is computed. This matrix helps to understand how much the features vary from one another.

Eigenvalues and eigenvectors are calculated from the covariance matrix.

After calculating eigenvalues and eigenvectors, I sorted eigenvalues in descending order, and the top n_components eigenvectors are selected.

The original data is projected onto the new subspace which I defined by the principal components

The same transformation is applied to the test dataset after centering it

```python
def apply_pca(X, n_components):
    X_mean = np.mean(X, axis=0)
    X_centered = X - X_mean  # Center the data
    covariance_matrix = np.cov(X_centered.T)  # Compute covariance matrix
    eigenvalues, eigenvectors = np.linalg.eig(covariance_matrix)  # Compute Eigenvalues and Eigenvectors
    
    # Sort the eigenvalues and eigenvectors in descending order
    sorted_indices = np.argsort(eigenvalues)[::-1]
    eigenvalues = eigenvalues[sorted_indices]
    eigenvectors = eigenvectors[:, sorted_indices]
    
    principal_components = eigenvectors[:, :n_components]
    
    # Transform the data onto the new subspace
    X_pca = np.dot(X_centered, principal_components)
    return X_pca, principal_components, X_mean

n_components = 2
X_train_pca, principal_components, X_train_mean = apply_pca(X_train_normalized, n_components)

X_test_centered = X_test_normalized - X_train_mean
X_test_pca = np.dot(X_test_centered, principal_components)

X_test_pca
```

As it seems, PCA works correctly. I want to plot it to understand more of the data.

```python
plt.figure(figsize=(8, 6))

# Scatter plot for training data
plt.scatter(X_train_pca[:, 0], X_train_pca[:, 1], c=y_train, cmap='viridis', s=50, alpha=0.7, edgecolors='k', label='Train Data')
    
# Scatter plot for test data 
plt.scatter(X_test_pca[:, 0], X_test_pca[:, 1], c=y_test, cmap='coolwarm', s=50, alpha=0.7, edgecolors='k', label='Test Data')

plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Dataset (2 Components)')
plt.legend(loc='best')
plt.grid(True)
plt.show()
```
